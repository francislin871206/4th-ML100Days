- 以下程式碼將示範在 python 如何利用 numpy 計算出兩組數據之間的相關係數，並觀察散佈圖
- 藉由觀察相關矩陣與散佈圖的關係, 希望同學對 負相關 的變數分布情形也有比較直覺的理解

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

# 弱相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
np.random.seed(1)
x = np.random.randint(0,50,1000)
np.random.seed(2)
y = np.random.randint(0,50,1000)
## 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x,y)
```

```python
# 將分布畫出來看看吧
plt.scatter(x,y)
```

# 正相關

```python
# 這次讓 y 與 x 正相關，再增加一些雜訊
for i in range(10000):
    x = np.random.randint(0,50,1000)
    y = x + np.random.normal(0,10,1000)
    co = np.corrcoef(x,y)
    print(co)
    if co[1][0] > 0.5:
        plt.scatter(x,y)
        plt.show()
        break
```

# 負相關

```python
for i in range(1000):
    x = np.random.randint(0,50,1000)
    y = 50 - x - np.random.normal(0,10,1000) 
    co = np.corrcoef(x,y)
    print(co)
    if co[1][0] < -0.5:
        plt.scatter(x,y)
        plt.show()
        break
```
