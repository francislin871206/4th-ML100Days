```python
#擅長資料格式的調整
import pandas as pd
#常用於陣列之間的運算
import numpy as np
#擅長繪圖
import matplotlib.pyplot as plt 
```

```python
#y為預測值,yp為實際觀測值
#計算均方誤差
#將兩個陣列相減後取絕對值，再用sum()將整個陣列加在一起
def mean_absolute_error(y,yp):
    mae = sum(abs(y-yp)/len(y))
    return mae

#計算平均絕對誤差
def mean_square_error(x,xp):
    mse = sum(((x-xp)*(x-xp))/len(x))
    return mse
```

```python
w = 5
b = 0.5
X_lin = np.linspace(0,100,101)
y = (X_lin + np.random.randn(101)*5)*w+b
#開始繪圖，第一個為x座標，第二個為y座標，第三個為資料呈現的型態和顏色，第四個為Label的名稱)
plt.plot(X_lin, y, 'b.', label='data points')
#圖表的名字
plt.title('Assume we have data points')
#label顯示的位置
plt.legend(loc=2)
#繪圖
plt.show()
```

```python
#製作沒有隨機變數的預期線圖
y_hat = X_lin * w + b
y = (X_lin + np.random.randn(101) * 5) * w + b
plt.plot(X_lin, y, 'b.', label='data')
plt.plot(X_lin, y_hat, 'r-', label='prediction')
plt.title('Assumn we have data points(and data prediction)')
plt.legend(loc=2)
plt.show()
```

```python
#計算平均絕對誤差
MAE = mean_absolute_error(y,y_hat)
#計算平均平方誤差
MSE = mean_square_error(y,y_hat)
print('The mean absolute error is %.3f' %(MAE))
print('The mean square error is %.4f' %(MSE))
```
